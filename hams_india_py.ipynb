```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
hams_india=pd.read_excel('Desktop\\empiezoitsolutions\\hams\\hams_india.xlsx')
```

```python
hams_india
```

```python
hams_india.drop(index=hams_india.index[-1], 
        axis=0, 
        inplace=True)
```

```python
hams_india
```

```python
hams_india.isna().sum()
```

```python
hams_india['Licences to be issued  (January to March)'].fillna(round(hams_india['Licences to be issued  (January to March)'].mean()),inplace=True)
```

```python
hams_india.isna().sum()
```

```python
hams_india['Renewals to be issued(January to March)'].fillna(round(hams_india['Renewals to be issued(January to March)'].mean()),inplace=True)
```

```python
hams_india.isna().sum()
```

```python
hams_india.rename(columns = {'Start-Year':'Year'}, inplace = True)
```

```python
hams_india
```

```python
hams_india.dtypes
```

```python
ax = hams_india.plot(x='Start -Year', y='Licences issued(April to December)',legend=False)
ax2 = ax.twinx()
hams_india.plot(x='Start -Year', y='Licences to be issued  (January to March)', ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```

```python
ax = hams_india.plot(x='Start -Year', y='Renweals issued (April to December)',legend=False)
ax2 = ax.twinx()
hams_india.plot(x='Start -Year', y='Renewals to be issued(January to March)', ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```

```python
corr=hams_india.corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True)
plt.title("Correlation map for entire dataset")
```

```python
highest_licence_issued=hams_india.groupby('Start -Year')['Licences issued(April to December)'].max().astype(int).reset_index()
highest_licence_issued.sort_values(by ='Licences issued(April to December)',ascending=False,inplace=True)
highest_licence_issued
```

```python
highest_licence_to_be_issued=hams_india.groupby('Start -Year')['Licences to be issued  (January to March)'].max().astype(int).reset_index()
highest_licence_to_be_issued.sort_values(by ='Licences to be issued  (January to March)',ascending=False,inplace=True)
highest_licence_issued
```

```python
highest_licence_issued_comb=hams_india.groupby('Start -Year')[['Licences issued(April to December)','Licences to be issued  (January to March)']].max().astype(int).reset_index()
highest_licence_issued_comb.sort_values(by =['Licences issued(April to December)','Licences to be issued  (January to March)'],ascending=[False,False],inplace=True)
highest_licence_issued_comb
```

```python
highest_renewal_issued=hams_india.groupby('Start -Year')['Renweals issued (April to December)'].max().astype(int).reset_index()
highest_renewal_issued.sort_values(by ='Renweals issued (April to December)',ascending=False,inplace=True)
highest_renewal_issued
```

```python
highest_renewal_to_be_issued=hams_india.groupby('Start -Year')['Renewals to be issued(January to March)'].max().astype(int).reset_index()
highest_renewal_to_be_issued.sort_values(by ='Renewals to be issued(January to March)',ascending=False,inplace=True)
highest_renewal_to_be_issued
```

```python
highest_renewal_issued_comb=hams_india.groupby('Start -Year')[['Renweals issued (April to December)','Renewals to be issued(January to March)']].max().astype(int).reset_index()
highest_renewal_issued_comb.sort_values(by =['Renweals issued (April to December)','Renewals to be issued(January to March)'],ascending=[False,False],inplace=True)
highest_renewal_issued_comb
```
